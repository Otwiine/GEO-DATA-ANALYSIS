```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx
import leafmap
```

```python
file_path = '/Users/otwiine/Desktop/gadm41_USA_shp/gadm41_USA_1.shp'
```

```python
USA = gpd.read_file(file_path)
```

```python
USA = USA.to_crs(epsg=4326)
```

```python
USA.explore()
```

```python
US = leafmap.Map()
US.add_basemap()
US.add_shp('/Users/otwiine/Desktop/gadm41_USA_shp/gadm41_USA_1.shp')
```

```python
ax = USA.centroid.plot()
ctx.add_basemap(ax, crs=USA.crs.to_string())
ax.axis("off")
plt.title("Centroids of States of the United states of America")
```
